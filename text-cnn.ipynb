```python
from sklearn.model_selection import train_test_split
import pandas as pd
import jieba
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
pd.options.display.max_rows=10
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./dataset/SMSSpamCollection',sep='\t',encoding='ISO-8859-1',names=['label','messages'])
```

```python
df['label']=df.label.map({'ham':0,'spam':1})
df.head()
```

```python
cw=lambda x:list(jieba.cut(x))
df['words']=df['messages'].apply(cw)
```

```python
tokenizer=Tokenizer()
tokenizer.fit_on_texts(df['words'])
vocab=tokenizer.word_index
vocab
```

```python
 x_train, x_test, y_train, y_test = train_test_split(df['words'], df['label'], test_size=0.1)
```

```python
print('Number of rows in the total set:{}'.format(df.shape[0]))
print('Number of rows in the training set:{}'.format(x_train.shape[0]))
print('Number of rows in the test set: {}'.format(x_test.shape[0]))
```

```python
x_train_word_ids=tokenizer.texts_to_sequences(x_train)
x_test_word_ids = tokenizer.texts_to_sequences(x_test)
```

```python
x_train_padded_seqs=pad_sequences(x_train_word_ids,maxlen=50) 
x_test_padded_seqs=pad_sequences(x_test_word_ids, maxlen=50)
x_test_padded_seqs
```
